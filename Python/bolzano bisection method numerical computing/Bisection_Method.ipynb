# For Complex Equation 

```
print("f(x) = x-np.exp(1/x)\n")

def f(x):
    return (x-np.exp(1/x))

def bisection_method(a, b, tol):
    n = 1
    tab = np.empty((0,5), dtype=float)
   

  
    if f(a)*f(b) > 0:
        #end function, no root.
        print("No root found.")

    else:
        while (b - a)/2.0 > tol:
         

            midpoint = (a + b)/2.0
            fx = float(f((round(midpoint,4)))) 
            fx_fix = round(fx,4)
            tab = np.append(tab,np.array([[n,round(b,4),round(a,4),round(midpoint,4),('{0:.4f}'.format(fx_fix))]]), axis =0, )

            if f(midpoint) == 0:
                n=n+1
                return(midpoint) #The midpoint is the x-intercept/root.
            elif f(a)*f(midpoint) < 0: # Increasing but below 0 case
                b = midpoint
                n=n+1

            else:
                a = midpoint
                n=n+1

        
        return(midpoint,n,tab)
answer,n, tab= bisection_method(1, 2, 0.0001)
print("N:",n, "Answer: ",  round(answer, 4))
df = pd.DataFrame(data=tab, columns=["N", "A (-ve)", "B (+ve)", "MidPoint", "f(x)"])
df.index = np.arange(1,len(df)+1)
df
```
